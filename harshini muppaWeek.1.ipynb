# DATA CLEANING

```python
columns_to_drop = ['EmployeeCount', 'EmployeeNumber', 'StandardHours']
data= df.drop(columns=columns_to_drop)
data
```

```python
data.head()
```

```python
duplicates = data.duplicated().sum()
print("Number of duplicate rows:", duplicates)
```

```python
def identify_outliers(df):
    outliers = {}
    for column in df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[column] = df[(df[column] < lower_bound) | (df[column] > upper_bound)].shape[0]
    return outliers

outliers = identify_outliers(data)
print(outliers)
```

# DATA ENCODING

```python
lable = LabelEncoder()
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns
data_encoded = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
```

```python
for column in categorical_columns:
    data[column] = lable.fit_transform(data[column])
```
