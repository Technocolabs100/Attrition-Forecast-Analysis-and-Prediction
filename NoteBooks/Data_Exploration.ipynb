```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

### 2.1 Attrition

```python
attrition_counts = df['Attrition'].value_counts()

plt.figure(figsize=(5, 3))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])
plt.title('Attrition Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

### 2.2 Age

```python
plt.figure(figsize=(5 , 3))
plt.hist(df['Age'] ,bins=10 , color='skyblue' , edgecolor = 'black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### 2.3 Monthly Income vs Attrition 

```python
plt.figure(figsize=(5,3))
df.boxplot(column='MonthlyIncome' , by='Attrition' ,grid=False , patch_artist=True, showfliers=False )
plt.title('Monthly Income vs. Attrition')
plt.suptitle('')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

### 2.4 Department vs. Attrition

```python
department_attrition = df[df['Attrition'] == 'Yes']['Department'].value_counts()

plt.figure(figsize=(6, 4))
department_attrition.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Department vs. Attrition')
plt.xlabel('Department')
plt.ylabel('Number of Attrition Cases')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```
